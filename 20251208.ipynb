```python
import json
import urllib.request as req

url = "https://api.gamer.com.tw/anime/v1/danmu.php?videoSn=36632&geo=TW%2CHK"
f = req.urlopen(url)
# content = f.read()
# print(content)
content_convert = json.loads(content)
# print(type(content_convert))

f = open("danmu.csv", "w", encoding="utf-8")
f.write("userid,text\n")

danmus = content_convert["data"]["danmu"]
for d in danmus:
  line = d["userid"] + ',"' + d["text"] + '"\n'
  f.write(line)
f.close()
```
