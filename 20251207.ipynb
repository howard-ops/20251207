```python
import random

def lottary_creat(low=1, high=49, count=6):

  lottary = set()
  while True:
    n = random.randint(low, high)
    lottary.add(n)
    if len(lottary) == count:
      break
  return lottary

prize = lottary_creat()
#特別號 in
special = None
while True:
  a = random.randint(1, 49)
  if not a in prize:
    special = a
    break
print(prize, special)

#一直買彩券 直到二等(11) 三等(10) 看買幾張
count = 0
score = 0
while True:
  ticket = lottary_creat()
  score = len(ticket & prize) * 2
  if special in ticket:
    score = score + 1
  count = count + 1
  if score >= 11:
    break
print(count, score)
```

```python
# 如果你想放到set裡或者當作dict key
# 這個東西一定不能可以變長/變短/修改(不可變動unmutable)
# !!!! tuple
import random
import datetime as dt

count = 0
birthdays = set()
while True:
    n = random.randint(1, 365)
    count = count + 1
    date_rand = "2025-" + str(n).zfill(3)
    date_convert = dt.datetime.strptime(date_rand, "%Y-%j")
    d = (date_convert.month, date_convert.day)
    if d in birthdays:
        print("幾個人才有同月同日生:", count)
        break
    else:
        birthdays.add(d)
        print(birthdays)
```

```python
# 10個人有同樣生日的機率
# 1 - 10個人都不一樣的機率
total = 1
start = 365
p = 1
for i in range(10):
    p = p * (start / 365)
    start = start - 1
print(total - p)
```
