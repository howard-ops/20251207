```python
import random

def lottary_creat(low=1, high=49, count=6):

  lottary = set()
  while True:
    n = random.randint(low, high)
    lottary.add(n)
    if len(lottary) == count:
      break
  return lottary

prize = lottary_creat()
#特別號 in
special = None
while True:
  a = random.randint(1, 49)
  if not a in prize:
    special = a
    break
print(prize, special)
```
