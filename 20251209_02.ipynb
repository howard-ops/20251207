```python
# 字串專屬功能: split/join/replace/strip
# pandas: 1. DataFrame(表格型態, 有行有列)
#      2. Series(一行或者一列, 接近list)

def func(s):
  spl = s.split(" / ")
  if len(spl) >= 2:
    area = spl[0]
    food = spl[1]
  elif len(spl) == 1:
    area = "-"
    food = spl[0]
  else:
    area = "-"
    food = "-"
  return pd.Series({"地點":area, "種類":food})

df[["地點", "種類"]] = df["地點_種類"].apply(func)
df
```

```python
# 請幫我把定休日變成七個欄位
# open(星期一): Y/N
def func(s):
    row = {
        "open_mon": "Y",
        "open_tue": "Y",
        "open_wed": "Y",
        "open_thu": "Y",
        "open_fri": "Y",
        "open_sat": "Y",
        "open_sun": "Y"
    }
    if "星期一" in s:
        row["open_mon"] = "N"
    if "星期二" in s:
        row["open_tue"] = "N"
    if "星期三" in s:
        row["open_wed"] = "N"
    if "星期四" in s:
        row["open_thu"] = "N"
    if "星期五" in s:
        row["open_fri"] = "N"
    if "星期六" in s:
        row["open_sat"] = "N"
    if "星期日" in s:
        row["open_sun"] = "N"
    return pd.Series(row)

# df[["open_mon", "open_tue", "open_wed", "open_thu", "open_fri", "open_sat", "open_sun"]] = df["定休日"].apply(func)
table2 = df["定休日"].apply(func)
df = pd.concat([df, table2], axis=1)
df = df.drop("定休日", axis=1)
df
```

```python
# and: & or: |
fil1 = (df["open_mon"] == "Y")
fil2 = (df["open_tue"] == "Y")
fil_day = fil1 | fil2

fil3 = df["種類"].str.contains("冰淇淋")
fil4 = df["種類"].str.contains("馬卡龍")
fil_food = fil3 | fil4
# def func(s):
#     return "冰淇淋" in s
# df["種類"].apply(func)

fil_all = fil_food & fil_day
df[fil_all]
```
