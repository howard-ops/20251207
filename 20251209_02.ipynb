```python
# 字串專屬功能: split/join/replace/strip
# pandas: 1. DataFrame(表格型態, 有行有列)
#      2. Series(一行或者一列, 接近list)

def func(s):
  spl = s.split(" / ")
  if len(spl) >= 2:
    area = spl[0]
    food = spl[1]
  elif len(spl) == 1:
    area = "-"
    food = spl[0]
  else:
    area = "-"
    food = "-"
  return pd.Series({"地點":area, "種類":food})

df[["地點", "種類"]] = df["地點_種類"].apply(func)
df
```
